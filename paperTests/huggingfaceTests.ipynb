```python
from huggingface_hub import HfApi

api = HfApi()
dataset = "zkdeng/spiderTraining1000-1000"

# List all files in the dataset repository
files = api.list_repo_files(repo_id=dataset, repo_type="dataset")

# Filter for parquet files
parquet_files = [f for f in files if f.endswith('.parquet')]
print("Parquet files:")
for file in parquet_files:
    print(f"  {file}")
```

```python
from huggingface_hub import HfApi

api = HfApi()

# Get detailed file info including sizes and dates
repo_info = api.list_repo_tree(repo_id=dataset, repo_type="dataset", recursive=True)

parquet_files = []
for item in repo_info:
    if hasattr(item, 'path') and item.path.endswith('.parquet'):
        parquet_files.append({
            'path': item.path,
            'size': getattr(item, 'size', None),
            'last_modified': getattr(item, 'last_modified', None)
        })

# Display results
import pandas as pd
df = pd.DataFrame(parquet_files)
display(df)
```

```python
from huggingface_hub import HfApi, hf_hub_url

api = HfApi()

# Get parquet files
files = api.list_repo_files(repo_id=dataset, repo_type="dataset")
parquet_files = [f for f in files if f.endswith('.parquet')]

# Get download URLs
file_urls = []
for file in parquet_files:
    url = hf_hub_url(repo_id=dataset, filename=file, repo_type="dataset")
    file_urls.append({
        'filename': file,
        'url': url
    })

# Display as DataFrame
import pandas as pd
df = pd.DataFrame(file_urls)
display(df)
```

```python
import polars as pl
import os

# Assuming 'df' is your polars DataFrame and you want row 0
row = df[0]

# Extract the image struct
image_struct = row['image'][0]

# Get the binary data and filename
image_bytes = image_struct['bytes']  # The binary data
filename = image_struct['path']     # The filename (e.g., "102535003.jpg")

# Get the label (subfolder name)
label = str(row['label'][0])

# Create subfolder
subfolder_path = f"images/{label}"
os.makedirs(subfolder_path, exist_ok=True)

# Save the image
output_path = os.path.join(subfolder_path, filename)
with open(output_path, 'wb') as f:
    f.write(image_bytes)

print(f"Saved image to: {output_path}")
```
